# Modeling Alluvial Channel Evolution

### Overview
In class, we discussed an equation that can be used to determine the unit sediment flux ($q_s$) in an alluvial channel. The equation is given by

$$q_s=\kappa(\tau-\tau_c)^{3/2}$$

where $\tau=\rho g h S$, $\tau_c$ is the critical shear stress required for sediment transport, and $\kappa$ is an empirical coeficient. This is a powerful relationship, but it does not directly tell us anything about how elevation will change with time at a particular location on the landscape. The quantify changes in elevation with time, we need to compute the derivative of the sediment flux in the along-channel direction. Fundamentally, our reason for doing this is that changes in elevation are related to spatial variations in sediment flux (change in elevation = flux in - flux out). Therefore, the model that we use to determine changes in elevation ($z$) with time ($t$) can be written as

$$ \frac{\partial z}{\partial t}=\kappa\frac{\partial q_s}{\partial x}$$

Below, we use this equation to simulate how an alluvial channel will respond to a rapid change in base level.

To visualize the scenario that we will consider, examine the figure above. It shows a high-order channel in blue as well as a low order tributary (red). Imagine that the high-order channel rapidly incises by 5 m over a short period of time. The direct result of this rapid incising is that its tributary (red) will experience a sudden 5 m drop in base level. **Run the code block below, which is set up to plot the longitudinal profile of the tributary channel immediately before and immediately after the drop in base level.**



```python
time=0.01                          # Time [kyr]

# !! YOU DO NOT NEED TO MODIFY THE CODE BELOW THIS LINE !! 
kappa=20                          # Erodibility coefficient
deltaz=5                          # Base level drop [m]
uplift=0.02                       # Uplift rate [m/kyr]

import numpy as np
from alluvialchannelevolution import alluvialchannel
[distance,area,elevationi,slope]=alluvialchannel(time,uplift,kappa,kappa,deltaz)

itopo=np.loadtxt('steadystatealluvial_U0pt02.txt')

import matplotlib.pyplot as plt
plt.rc('font', size=18)          # controls default text sizes
fig = plt.figure(figsize=(15, 8))
# Adds subplot on position 1
ax1 = fig.add_subplot(121)
# Adds subplot on position 2
ax2 = fig.add_subplot(122)
fig.tight_layout(pad=4.0)
ax1.plot(distance,elevationi,'k-',label="Modeled Topography")
ax1.plot(distance,itopo,'r--',label="Initial Topography")
ax1.legend()
ax1.set(xlabel='Distance (m)', ylabel='Elevation (m)', title='t = '+str(round(time,2)))
ax2.plot(distance,elevationi,'k-',label="Modeled Topography")
ax2.plot(distance,itopo,'r--',label="Initial Topography")
ax2.set_xlim([1500, 2000])
ax2.set_ylim([-5, 10])
ax2.legend()
ax2.set(xlabel='Distance (m)', ylabel='Elevation (m)', title='Close-up view at t = '+str(round(time,2)))
plt.show()
```

The figures that are generated by the model show the 5 m drop in elevation where the tributary meets the main channel. How will the elevation in the tributary change in response to this drop in base level? **In the code block below, change the simulation time from 0.01 kyr to 1 kyr to see a model prediction for channel profile after 1 kyr of time has passed.**

```python
time=1                            # Time [kyr]

# !! YOU DO NOT NEED TO MODIFY THE CODE BELOW THIS LINE !! 
kappa=20                          # Erodibility coefficient
deltaz=5                          # Base level drop [m]
uplift=0.02                       # Uplift rate [m/kyr]

import numpy as np
from alluvialchannelevolution import alluvialchannel
[distance,area,elevation1,slope]=alluvialchannel(time,uplift,kappa,kappa,deltaz)

itopo=np.loadtxt('steadystatealluvial_U0pt02.txt')

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 9))
plt.rc('font', size=18)          # controls default text sizes
plt.plot(distance,elevationi,'k-',label="t=0.01 kyr")
plt.plot(distance,elevation1,'c-',label="t=1 kyr")
plt.plot(distance,itopo,'r--',label="Initial Topography")
plt.xlabel('Distance (m)', fontsize=18)
plt.ylabel('Elevation (m)', fontsize=18)
plt.title('t = '+str(round(time,1)), fontsize=18)
plt.legend()
plt.rc('legend', fontsize=18)    # legend fontsize
plt.show()
```

After 1000 kyr of time has passed following the change in base level, the channel has adjusted everywhere to be approximately 5 m lower in elevation. In other words, the overall shape of **the channel profile looks exactly the same as it did before the drop in base level (it is just shifted down by 5 m)**. It has reached a new "steady state" and will continue to maintain this same shape unless some other change in base level, climate, or uplift occurs.
